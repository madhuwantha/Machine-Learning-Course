```python
import pandas as pd
```

```python
#crreating a series of car types
cars = pd.Series(["BMW","Toyota","Honda"])
cars
```

```python
#creating series of colors
colours = pd.Series(["Blue","Red","White"])
colours
```

```python
#creating DataFrame of car with colors
car_data = pd.DataFrame({
    "Car_type":cars,
    "Colour"  : colours
})
```

```python
car_data
```

```python
#creating series of foods
foods = pd.Series(["Lemon","Chicken","Rice"])
#creating series of dollar
prices = pd.Series([5,6,88])
```

```python
#creating DataFrame of food with priccess
food_data = pd.DataFrame({
    "food":foods,
    "price":prices
})
food_data
```
