```python
# Standard imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# 0. An end-to-end Scikit-Learn workflow

```python
to# 1. Get the data ready
heart_disease = pd.read_csv("Data/Scikit-Learn/heart-disease.csv")
```

```python
heart_disease
```

```python
# Create X (features matrix)
x = heart_disease.drop("target",axis=1)

# Create y (labels)
y = heart_disease["target"]
```

```python
# 2. Choose the right model and hyperparameters
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()

# We'll keep the default hyperparameters
clf.get_params()
```

```python
# 3. Fit the model to the training data
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
clf.fit(x_train,y_train)
```

```python
len(heart_disease), len(x_test),len(x_train)
```

```python
# make a prediction
y_label = clf.predict(np.array([63,1,3,145,233,1,0,150,0,2.3,0,0,1]))
```

```python
y_preds = clf.predict(x_test)
y_preds
```

```python
y_test
```

```python
# 4. Evaluate the model on the training data and test data
clf.score(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(classification_report(y_test,y_preds))
```

```python
confusion_matrix(y_test,y_preds)
```

```python
accuracy_score(y_test,y_preds)
```

```python
# 5. Improve a model
# Try different amount of n_estimators
np.random.seed(42)
for i in range(10,100,10):
    print(f"Trying model with {i} estimators...")
    clf = RandomForestClassifier(n_estimators=i).fit(x_train,y_train)
    print(f"Model accuracy on test set: {clf.score(x_test, y_test) * 100:.2f}%")
    print("")
```

```python
# 6. Save a model and load it
import pickle

pickle.dump(clf,open("Data/Scikit-Learn/random_forst_model_1.pkl", "wb"))
```

```python
loaded_model = pickle.load(open("Data/Scikit-Learn/random_forst_model_1.pkl", "rb"))
loaded_model.score(x_test,y_test)
```
